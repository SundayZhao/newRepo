```python
import random
#从datasets_generate.ipynb中生成的数据集文件名
#输入文件
buggy_code_file='.\java.buggy'
fixed_code_file='.\java.fixed'
#输出文件
#训练集
output_train_buggy=open(r'.\train.buggy-fixed.buggy','w',encoding='utf-8')
output_train_fixed=open(r'.\train.buggy-fixed.fixed','w',encoding='utf-8')
#验证集
output_valid_buggy=open(r'.\valid.buggy-fixed.buggy','w',encoding='utf-8')
output_valid_fixed=open(r'.\valid.buggy-fixed.fixed','w',encoding='utf-8')
#测试集
output_test_buggy=open(r'.\test.buggy-fixed.buggy','w',encoding='utf-8')
output_test_fixed=open(r'.\test.buggy-fixed.fixed','w',encoding='utf-8')
#拆分参数
training_proportion=0.8
validation_proportion=0.2
test_proportion=0
assert training_proportion + validation_proportion+test_proportion==1
assert training_proportion >0
assert validation_proportion >0

buggy_code_file=open(buggy_code_file,'r',encoding='utf-8')
fixed_code_file=open(fixed_code_file,'r',encoding='utf-8')
buggy_codes=buggy_code_file.readlines()
fixed_codes=fixed_code_file.readlines()
all_codes=list(map(lambda x, y:(x,y), buggy_codes, fixed_codes))
buggy_codes=[]
fixed_codes=[]
random.shuffle(all_codes)
code_len=len(all_codes)
training_array=all_codes[:int(len(all_codes)*training_proportion)]
validation_array=[]
test_array=[]
if training_proportion +validation_proportion==1:
    validation_array=all_codes[int(code_len*training_proportion)+1:]
else:
    validation_array=all_codes[int(code_len*training_proportion)+1:int(code_len*(training_proportion+validation_proportion))]

if test_proportion>0:
    test_array=all_codes[int(code_len*(training_proportion+validation_proportion))+1:]

all_codes=[]
#输出数据集
for (buggy,fixed) in training_array:
    output_train_buggy.write(buggy+'\n')
    output_train_fixed.write(fixed+'\n')
for (buggy,fixed) in validation_array:
    output_valid_buggy.write(buggy+'\n')
    output_valid_fixed.write(fixed+'\n')
for (buggy,fixed) in test_array:
    output_test_buggy.write(buggy+'\n')
    output_test_fixed.write(fixed+'\n')

output_train_buggy.close()
output_train_fixed.close()
output_valid_buggy.close()
output_valid_fixed.close()
output_test_buggy.close()
output_test_fixed.close()
buggy_code_file.close()
fixed_code_file.close()
print('finish')
print('train set:',len(training_array))
print('valid set:',len(validation_array))
print('test set:',len(test_array))
```
